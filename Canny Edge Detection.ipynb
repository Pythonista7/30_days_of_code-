# Canny Edge Detection


The Canny edge detector is an edge detection operator that uses a multi-stage algorithm to detect a wide range of edges in images. It was developed by John F. Canny in 1986. 

The Process of Canny edge detection algorithm can be broken down to 5 different steps:

1. Apply <b>Gaussian filter to smooth the image in order to remove the noise</b>

2. <b>Find the intensity gradients</b> of the image

3. Apply <b>non-maximum suppression to get rid of spurious response to edge detection</b>

4. Apply <b>double threshold to determine potential edges</b>

5. <b>Track edge by hysteresis</b>: Finalize the detection of edges by suppressing all the other edges that are weak      and not connected to strong edges.

For step-by-step explaination visit [here](https://docs.opencv.org/3.1.0/da/d22/tutorial_py_canny.html)

**OpenCV puts all the above in single function, cv2.Canny().**

First argument is our input image. Second and third arguments are our minVal and maxVal respectively. Third argument is aperture_size. It is the size of Sobel kernel used for find image gradients. By default it is 3. Last argument is L2gradient which specifies the equation for finding gradient magnitude.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('assets/skyscraper.jpeg')
plt.imshow(img,cmap='gray')
```

```python
edges=cv2.Canny(img,100,150)
plt.imshow(edges,cmap='gray')
```
